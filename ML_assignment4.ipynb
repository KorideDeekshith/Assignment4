# 2nd question

```python
Kmean = pd.read_csv("C:/Users/deeks/Downloads/datasets(1)/datasets/K-Mean_Dataset.csv")
```

```python
Kmean
```

```python

A = Kmean.iloc[:,1:].values

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(A)
A = imputer.transform(A)
```

```python

wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(A)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('the elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('Wcss')
plt.show()
```

```python
from sklearn.cluster import KMeans
nclusters = 4 
km = KMeans(n_clusters=nclusters)
km.fit(A)
```

```python
KMeans
```

```python

B_cluster_kmeans = km.predict(A)
from sklearn import metrics
score = metrics.silhouette_score(A, B_cluster_kmeans)
print('Silhouette score:',score)
```
